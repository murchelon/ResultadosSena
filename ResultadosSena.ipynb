```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("csv/mega_sena_asloterias_ate_concurso_2146_sorteio.csv", sep=";")

df["Soma"] = df.B1 + df.B2 + df.B3 + df.B4 + df.B5 + df.B6
# df


reg = linear_model.LinearRegression()

reg.fit(df[["CONCURSO"]], df.Soma)


reg.predict([[200]])

%matplotlib inline
plt.figure(figsize=(14,6))
plt.xlabel("Concurso")
plt.ylabel("Soma")
plt.scatter(df.CONCURSO, df.Soma, linewidths=0.1, marker="o")

plt.plot(df.CONCURSO], reg.predict([[200]]), color="red")
```
